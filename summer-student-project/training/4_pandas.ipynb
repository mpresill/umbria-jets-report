# Introduction to Pandas

Pandas is a Python package providing fast, flexible, and expressive data structures designed to make working with "relational" or "labeled" data both easy and intuitive. It aims to be the fundamental high-level building block for doing practical, real-world data analysis in Python.

We often have our data in pandas dataframes, only converting it to tensors or arrays when necessary, to improve interpretability and to simplify working with multidimensional data. 

pandas is built on top of numpy and is intended to integrate well within a scientific computing environment.

## Getting Started

The basic building blocks of the pandas library are `Series`, `DataFrame`, and `Index`. In this section, you will learn how to create them and perform basic operations.

We will first import the package. The usual abbreviation is `pd`.

```python
import pandas as pd
```

A **DataFrame** is composed of one or more **Series**. The names of the **Series** form the column names, and the row labels form the **Index**. The data we use here is the Titanic dataset, stored as a CSV file. The data consists of the following columns:

- `PassengerId`: Id of every passenger.
- `Survived`: Indication whether passenger survived. 0 for yes and 1 for no.
- `Pclass`: One out of the 3 ticket classes: Class 1, Class 2 and Class 3.
- `Name`: Name of passenger.
- `Sex`: Gender of passenger.
- `Age`: Age of passenger in years.
- `SibSp`: Number of siblings or spouses aboard.
- `Parch`: Number of parents or children aboard.
- `Ticket`: Ticket number of passenger.
- `Fare`: Indicating the fare.
- `Cabin`: Cabin number of passenger.
- `Embarked`: Port of embarkation.

```python
titanic = pd.read_csv("titanic.csv")
titanic
```

### Extracting Subsets

One of the main advantages and crucial parts of pandas is working with subsets: Choosing the part of the dataframe which is necessary for your calculations and performing computations only on that part of the dataframe.

We can extract subsets of both rows and columns.

#### Selecting Columns

We have already seen how we can select one column. If we select two or more columns, we have to use a list of column names inside the square brackets. This will return a dataframe instead of a Series, when we only had one column.

```python
titanic[["Age", "Sex"]]
```

We can get the shapes of dataframes and series with the `shape` attribute.

```python
titanic[["Age", "Sex"]].shape
```

#### Selecting Rows

We can select rows by the name of the index. Do not confuse this with the position of the index, the following selects the rows with the indices between [100, 110) and not those rows with positions [100, 110). In this dataframe, this is by chance the same.

```python
titanic[100:110]
```

## Indexing

We can index both with the names of rows and columns with the `loc` command or with the positions of rows and columns with `iloc`.

```python
titanic.loc[100:110, ["Age", "Sex"]]
```

```python
titanic.iloc[100:110, [4, 5]]
```

#### Masks

Additionally, we can filter with Boolean masks. We can define conditions on which entries in the dataframe are either `True` or `False` and only return those entries in which the condition is `True`.

Let's select all passengers over 35 years:

```python
age_mask = titanic["Age"] > 35
titanic[age_mask]
```

We can also do this with `loc`, which might be useful if we also have an index requirement or want to return only certain columns.

```python
titanic.loc[age_mask, ["Survived", "Name"]]
```

## Data Wrangling

Often, we want to work with the data given in our dataframes, e.g creating new columns, aggregating data, calculating difficult functions etc. 

#### Grouping

Let's start by grouping our data in distinctive groups. For example, we want to get different statistics for male and female passengers. For that, we can use the `groupby` function.

```python
titanic.groupby(by=['Sex'])
```

This returns a groupby object. With this we can first inspect the groups and then calculate all summary statistics on each group individually as well.

```python
titanic.groupby(by=['Sex']).groups
```
