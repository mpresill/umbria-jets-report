## Classes

To create a class, we'll use the `class` keyword. 
Lets's define a very basic class with property `x` and create an object of this one.

```python
class MyClass:
    x = 5

p1 = MyClass()
print(p1.x)
```

The `__init__()` function:

The examples above are classes and objects in their simplest form, and are not really useful in real-life applications.
To understand the meaning of classes, we have to understand the built-in `__init__()` function.
All classes have a function called `__init__()`, which is always executed when the class is being initiated.
Use the `__init__()` function to assign values to object properties, or other operations that are necessary to do when the object is being created. There are other build in methods, e.g. `__call__()`, can you find on google what this one is used for?

Define a `Person` class with the persons name and age as properties.

```python
class Person:
    def __init__(self, name, age):
      self.name = name
      self.age = age

p1 = Person("John", 36)

print(p1.name)
print(p1.age)
```

### Object Methods

Objects can also contain methods. Methods in objects are functions that belong to the object.
Let us create a method in the `Person` class.

```python
class Person:
    def __init__(self, name, age):
        self.name = name
        self.age = age

    def introduce(self):
        print('Hello my name is {} and I am {} years old'.format(self.name, self.age))

p1 = Person("John", 36)
p1.introduce() 
```

### The `self` parameter

The self parameter is a reference to the current instance of the class, and is used to access variables that belong to the class.

### Inheritance

A class can inherit properties from other classes by stating it in the brackets at its definition. Let's create a `Physicist` class that inherits properties from `Person` by stating it in brackets in their first initialization line.

```python
class Physicist(Person):
    def __init__(self, name, age, field):
        super().__init__(name, age)
        self.field = field
        self.favorite_number = 42
    
    def work(self):
        print("I love working in the field of " + self.field + ", my favorite number is {}".format(self.favorite_number))

p2 = Physicist('Chris', 19, 'particle physics')
p2.introduce()
p2.work()
```

Have an eye of the `super().__init__(name, age)` line: here we initialize the parent class (=`Person`) of `Physicist` with the needed arguments `name` and `age`.
You can also see that the object `p2` can access the `introduce()` method, since it is inherited from `Person`.
