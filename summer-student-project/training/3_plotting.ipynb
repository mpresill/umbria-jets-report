# Matplotlib

We often use visualization as way to understand and show our data and results. Understanding how plotting works and how to best visualize different data is important for every physicist, especially when dealing with large datasets. The `matplotlib` package provides easy plotting functions and many customization possibilities.

Usually, matplotlib is used through its `matplotlib.pyplot`, which is a collection of functions that make matplotlib work like MATLAB. 

This tutorial is based on:

https://matplotlib.org/stable/tutorials/

Matplotlib graphs your data on Figures (e.g., windows, Jupyter widgets, etc.), each of which can contain one or more Axes, an area where points can be specified in terms of $x$-$y$ coordinates (or $\theta$-$r$ in a polar plot, $x$-$y$-$z$ in a 3D plot, etc.). The simplest way of creating a Figure with an Axes is using pyplot.subplots. We can then use Axes.plot to draw some data on the Axes.

```python
import numpy as np
import matplotlib.pyplot as plt
```

## Basic Plots

The easiest way to visualize is with `plt.plot`. Here we generate sample data x and y, and then create a basic plot. We add a title, labels for the x and y axes and a legend. Finally, we can display the plot using `show()`.

```python
# Generating some sample data
x = np.linspace(0, 10, 100)
y = np.sin(x)

# Basic Plotting
plt.figure(figsize=(5, 3))
plt.plot(x, y, label='sin(x)')
plt.title('Basic Plot')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.show()
```

## Subplots

We often want to show several plots next or below each other, to make comparisons easier. For that, matplotlib offers the `subplot` possibility. We create one figure, but different axes objects which we can customize as we want.

In `plt.subplots` we first define the number of rows and the number of columns that we want to have on the figure. There are several other customization possibilities, which you can check out in the [documentation](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.subplots.html).

```python
fig, axs = plt.subplots(1, 2, figsize=(10, 5))
axs[0].plot(x, y, 'ro', label="Plot 1")
axs[1].plot(x, -y, '.g', label="Plot 2")

axs[0].legend()
axs[1].legend()
plt.show()
```

### 1D Histograms

As physicists, most of the time we need histograms. The `.hist` function works very similar to the `.plot` function. The good thing is that matplotlib automatically creates the bins and fills them with the correct entries, so we can simply give it our entire data and let it do its magic.

```python
def gaussian(x, mu, sig):
    return 1.0 / (np.sqrt(2.0 * np.pi) * sig) * np.exp(-np.power((x - mu) / sig, 2.0) / 2)

hist_data = np.random.normal(size=(1000,))

function = gaussian(np.linspace(-4, 4, 1000), 0.0, 1.0)

plt.figure(figsize=(7, 5))
n, bins, patches = plt.hist(hist_data, bins=100)
plt.show()
```

The function returns the values of the histogram bins (length = N$_{bins}$), the edges of the bins(length = N$_{bins}$ + 1) and the patches, which is the container of individual artists used to create the histograms. ### 2D Histograms

If we want to know the distribution of two-dimensional data, we can also plot 2D histograms. For 2D histograms, we use colormaps to show the amount of data in each bin. It is often a challenge to choose a good colormap.
Otherwise, 2D histograms work pretty similar to 1D histograms. One main difference is that it is very difficult to plot two distributions into one plot, as the plots will be very difficult to decipher. 

```python
x_2d = np.random.randn(1000)
y_2d = np.random.randn(1000)

plt.figure(figsize=(8, 6))
plt.hist2d(x_2d, y_2d, bins=(20, 20))
plt.title('2D Histogram')
plt.xlabel('X')
plt.ylabel('Y')
plt.colorbar(label='Frequency')
plt.show()
```

## Scatterplots

With standard plots and histograms we can show one- or two-dimensional data easily. If we increase the number of dimensions, good visualizations get increasingly more difficult. One possibility to show higher-dimensional data is a scatterplot. In a scatterplot, we show a two-dimensional space but can visualize a third dimension through color or size of the markers.

```python
x_scatter = np.random.rand(50)
y_scatter = np.random.rand(50)
colors = np.random.rand(50)

plt.figure(figsize=(7, 5))
plt.scatter(x_scatter, y_scatter, c=colors)
plt.title('Scatter Plot')
plt.xlabel('X')
plt.ylabel('Y')
plt.colorbar(label='Color Scale')
plt.show()
```
